```
#1次関数
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

x = np.arange(0,11)
y = 4*x+3
plt.plot(x,y)
plt.grid()
```

```
#連立方程式
from sympy import Symbol,solve
a = Symbol("a")
b = Symbol("b")
ex1 = 3*a + b -6
ex2 = 4*a + b -12
ans = solve((ex1,ex2))
print(ans[a],ans[b])
```
